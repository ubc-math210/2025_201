## Systems of DEs

Consider the equation

$$
my'' + cy' + ky = F(t)
$$

and $y(0) = 0$, $y'(0) = 0$. Write as a first order system $\mathbf{u}' = \mathbf{f}(t,\mathbf{u})$ where

$$
\mathbf{f}(t,\mathbf{u}) = \left( u_1 , \frac{1}{m}(F(t) - cu_1 - ku_0) \right)
$$

and $\mathbf{u} = (u_0,u_1)$.

```python
m = 1; c = 0.01; k = 2;
F = lambda t: np.cos(t)

f = lambda u,t: np.array([ u[1] , 1/m*(F(t) - c*u[1] - k*u[0]) ])
U0 = np.array([0.0,0.0])
t = np.linspace(1,50,1000)

U = spi.odeint(f,U0,t)
plt.figure(figsize=(10,3))
plt.plot(t,U[:,0])
plt.show()
```
