## Riemann Sums

Approximate the following integral using left and right Riemann sums for partitions with equally spaced points:

$$
\int_0^1 e^{-x^2} dx
$$

```python
x = np.linspace(-2,2,200)
y = np.exp(-x**2)
plt.figure(figsize=(10,3))
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
a = 0
b = 1
N = 50
dx = (b - a)/N
x = np.linspace(a,b,N+1)
y = np.exp(-x**2)
R = np.sum(y[1:]*dx)
print(R)
```

```python
L = np.sum(y[:-1]*dx)
print(L)
```
