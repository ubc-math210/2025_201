## Riemann Sum Error Formula

Approximate the definite integral with error less than $0.01$:

$$
\int_0^{\sqrt{\pi/2}} \sin(x^2) \, dx
$$

```python
f = lambda x: np.sin(x**2)
a = 0
b = np.sqrt(np.pi/2)
x = np.linspace(a,b,200)
y = f(x)
plt.plot(x,y), plt.grid(True)
plt.show()
```

```python
N = 197
dx = (b - a)/N
x = np.linspace(a,b,N + 1)
y = f(x)
I = np.sum(y[1:]*dx)
print(I)
```
