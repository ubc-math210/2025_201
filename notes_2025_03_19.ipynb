```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

## 2D Mass Spring System

$$
\begin{align*}
mx'' &= \frac{kLx}{\sqrt{x^2 + y^2}} - kx \\
my'' &= \frac{kLy}{\sqrt{x^2 + y^2}} - ky \\
\end{align*}
$$

```python
m = 1; L = 1; k = 0.2;

def f(u,t):
    D = L/np.sqrt(u[0]**2 + u[2]**2) - 1
    du0dt = u[1]
    du1dt = k*u[0]/m*D
    du2dt = u[3]
    du3dt = k*u[2]/m*D
    return np.array([du0dt,du1dt,du2dt,du3dt])

U0 = np.array([2,0,0,5])
t = np.linspace(0,200,1000)
U = spi.odeint(f,U0,t)

# zorder specifies how plots are layered in the same figure
# Higher zorder puts the red dot on top of the blue parametric cuurve
plt.plot(U[:,0],U[:,2],zorder=0)
plt.scatter(U0[0],U0[2],s=50,c='r',zorder=1)
plt.show()
```

The output `U` is a matrix with `len(t)` rows and `lengths(U0)` columns. Look at the first 5 rows:

```python
U[:5,:]
```

The coluumns correspond to the function $u_0(t) = x(t)$, $u_1(t) = x'(t)$, $u_2(t) = y(t)$, $u_3(t) = y'(t)$. The rows correspond to values at time $t_i$. Note that the top row at index 0 is exactly the initial values at time $t_0 = 0$.
