## Vectorization

Vectorization is the process of applying a function or operation to every entry in a vector elementwise. For example, let $\mathbf{x}$ be a vector

$$
\mathbf{x} = \begin{bmatrix} x_0 & x_1 & \cdots & x_N \end{bmatrix}
$$

Then a function $f(x)$ applied to the vector is

$$
f(\mathbf{x}) = \begin{bmatrix} f(x_0) & f(x_1) & \cdots & f(x_N) \end{bmatrix}
$$

We also have vectorized arithmetic operations

$$
\mathbf{x}^2 = \begin{bmatrix} x_0^2 & x_1^2 & \cdots & x_N^2 \end{bmatrix}
$$

$$
\mathbf{x} \mathbf{y} = \begin{bmatrix} x_0 y_0 & x_1 y_1 & \cdots & x_N y_N \end{bmatrix}
$$

Let $c$ be a scalar and compute

$$
\mathbf{x} + c = \begin{bmatrix} x_0 + c & x_1 + c & \cdots & x_N + c \end{bmatrix}
$$

We have seen vectorized operations in linear algebra already such as vector addition scalar multiplication:

$$
\mathbf{u} + \mathbf{v} = \begin{bmatrix} u_0 + v_0 & u_1 + v_1 & \cdots & u_N + v_N \end{bmatrix}
$$

$$
c \mathbf{u} = \begin{bmatrix} cu_0 & cu_1 & \cdots & cu_N \end{bmatrix}
$$

## Plotting Procedure

1. Create a vector of $x$ values.
2. Use vectorized operations to create vector of $y$ values.
3. Use `plt.plot(x,y)` then add style, title, labels, grids, etc.
4. Display the figure with `plt.show()`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Use NumPy function `np.linspace(a,b,N)` to create the vector of `N` evenly spaced values from `a` to `b` (inclusive). Note that the spacing between values is

$$
\frac{b - a}{N-1}
$$

```python
x = np.linspace(0,1,5)
print(x)
```

```python
x = np.linspace(0,1,101)
y = np.sin(2*np.pi*x)
plt.plot(x,y)
plt.title('Sine Function')
plt.xlabel('x values'), plt.ylabel('y values'), plt.grid(True)
plt.show()
```

## Examples: Solutions of ODEs

### Example 1

For each initial value $y(0) = 0.0,0.1,0.2,\dots,2.0$, plot the unique solution of the equation $y' = -y^2 t$.

$$
y' = -y^2 t
\ \Rightarrow \ -\int\frac{dy}{y^2} = \int t dt
\ \Rightarrow \ \frac{1}{y} = t^2/2 + C
$$

$$
y(t) = \frac{1}{t^2/2 + C}
$$

```python
t = np.linspace(0,5,101)
for y0 in np.arange(0.1,2.1,0.1):
    y = 1/(t**2/2 + 1/y0) # C = 1/y(0)
    plt.plot(t,y,'b')
plt.plot(t,0*t,'b')
plt.grid(True)
plt.show()
```

### Example 2

For each initial value $y'(0) = 0,0.1,\dots,1$, plot the unique solution of $y'' + 2y' + 2y = 0$, $y(0) = 0$.

```python
a = 1
b = 2
c = 2
D = b**2 - 4*a*c
D
```

Roots are complex $\alpha \pm i \beta$ and the general solution is

$$
y(t) = C_0 e^{\alpha t} \sin(\beta t) + C_1 e^{\alpha t} \cos(\beta t)
$$

If $y(0) = 0$ then $C_1 = 0$. Compute the derivative and plug in $t=0$ to get

$$
y'(0) = \beta C_0
$$

```python
alpha = -b/(2*a)
beta = abs(D)**0.5/(2*a)

t = np.linspace(0,8,101)
for dy0 in np.arange(0.0,1.1,0.1):
    y = dy0/beta*np.exp(alpha*t)*np.sin(beta*t)
    plt.plot(t,y,'b')
plt.grid(True)
plt.show()
```
